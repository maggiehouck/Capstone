```python
files = []
utah = []
wyoming = []
newmex = []
colorado = []
kansas = []
oklahoma = []

for filename in os.scandir(data_dir):
        if ".pdf" in filename.name:
            reader = pypdf.PdfReader(filename.path)
            pypdf2_text = ""
            for i in range(len(reader.pages)):
                pypdf2_text += reader.pages[i].extract_text()  
            files.append(filename.name)
            utah.append(get_courtcode(pypdf2_text, "Utah"))
            wyoming.append(get_courtcode(pypdf2_text, "Wyoming"))
            newmex.append(get_courtcode(pypdf2_text, "New Mexico"))
            colorado.append(get_courtcode(pypdf2_text, "Colorado"))
            kansas.append(get_courtcode(pypdf2_text, "Kansas"))
            oklahoma.append(get_courtcode(pypdf2_text, "Oklahoma"))
           
        else:
            continue
```

```python
df = pd.DataFrame({'file_name': files,
                      'Utah': utah,
                      'Wyoming': wyoming, 
                      'New Mexico': newmex, 
                      'Colorado': colorado, 
                      'Kansas': kansas, 
                      'Oklahoma': oklahoma})
```

```python
# define a function to find the column with the highest value for a row
def find_max_col(row):
    return row.idxmax()
```

```python
max_cols = df[['Utah', 'Wyoming', 'New Mexico', 'Colorado', 'Kansas', 'Oklahoma']].apply(find_max_col, axis=1)

df['state'] = max_cols
```

```python
df = df[['file_name', 'state']]
```

```python
# Get the unique values in the identifier column
identifiers = df['state'].unique()

# Create a subdirectory for each unique identifier
for identifier in identifiers:
    os.makedirs(identifier, exist_ok=True)
```

```python
# Move the files into the appropriate subdirectory based on their identifier
for index, row in df.iterrows():
    file_name = row['file_name']
    identifier = row['state']
    destination = os.path.join(identifier, file_name)
    shutil.move(file_name, destination)
```
